```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
data = pd.read_excel('26.4 Когортный анализ.xlsx')
```

```python
data.head()
```

```python
data.isnull().any()
```

```python
data.dropna(inplace=True)
```

```python
data.dtypes
```

```python
data.shape
```

```python
data.describe()
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month,1)
```

```python
data['month_year'] = data['date'].apply(get_month)
```

```python
data.head()
```

```python
data['cohortMonth'] = data.groupby('user_id')['month_year'].transform('min')
```

```python
data.head(50)
```

```python
def get_date(df,column):
    year=df[column].dt.year
    month=df[column].dt.month
    day=df[column].dt.day
    return year, month, day
```

```python
date_year,date_month,_ = get_date(data, 'date')
```

```python
cohort_year, cohort_month,_ = get_date(data, 'cohortMonth')
```

```python
month_diff = date_month-cohort_month
```

```python
year_diff = date_year-cohort_year
```

```python
data['cohort_index'] = year_diff*12+month_diff+1
```

```python
data.head()
```

```python
data['cohort_index'].unique()
```

```python
data1=data.pivot_table(index='cohortMonth',

columns='cohort_index',

values='user_id',

aggfunc=len)
```

```python
data1
```

```python
first_column = data1.iloc[:,0]
```

```python
retention = data1.div(first_column, axis=0)
```

```python
retention.round(2)
```

```python
plt.figure(figsize=(11, 9))

sns.heatmap(data=retention, annot=True, fmt='.0%', vmin=0.0, vmax=0.5, cmap='YlGnBu')

plt.show();
```

```python
data2=data.pivot_table(index='cohortMonth',

columns='cohort_index',

values='price',

aggfunc=sum)
```

```python
data2
```

```python
first_column = data2.iloc[:,0]
retention2 = data2.div(first_column, axis=0)
retention2.round(2)

```

```python

plt.figure(figsize=(11, 9))

sns.heatmap(data=retention2, annot=True, vmin=0.0, vmax=0.5, cmap='YlGnBu', fmt='.0%')

plt.show();
```
